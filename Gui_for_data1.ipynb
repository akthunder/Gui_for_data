```python
from tkinter import *
import pandas as pd
import matplotlib.pyplot as plt
```

```python
win = Tk()
win.geometry('500x700')
win.title('this is title')
line = StringVar()
scatter = StringVar()
bar  = StringVar()
circle= StringVar()
address= StringVar()
xcolumn= StringVar()
ycolumn= StringVar()
filename= StringVar()
extra2= StringVar()
col_val = IntVar()
row_val = IntVar()
null_val = IntVar()

def submit():
    if address.get():
        df1 = pd.read_csv(address.get())
        no_of_columns = df1.columns
        lenths =  {}
        for i in range(len(no_of_columns)):
            series = pd.Series({'lenths':(len(no_of_columns[i]))})
        series              
        max_of_string = series['lenths'].max()
        if max_of_string >20:
            print("please minimize your column name first \n")
        else:
            col_label = Label(win, text = no_of_columns[0],font = (10))
            col_label.place(x=150,y=200)
            col_label = Label(win, text = no_of_columns[1], font = (10))
            col_label.place(x=250,y=200)
            col_label = Label(win, text = no_of_columns[2], font = (10))
            col_label.place(x=350,y=200)
            col_label = Label(win, text = no_of_columns[3], font = (10))
            col_label.place(x=450,y=200)
            col_label = Label(win, text = no_of_columns[4], font = (10))
            col_label.place(x=550,y=200)
            col_label = Label(win, text = no_of_columns[5], font = (10))
            col_label.place(x=650,y=200)
            col_label = Label(win, text = no_of_columns[6], font = (10))
            col_label.place(x=750,y=200)
    else:
        print("fill the file location \n")
            

def line():
    if address.get() and xcolumn.get() and ycolumn.get(): 
        df = pd.read_csv(address.get())
        rows,columns = df.shape
        nul = df.isnull().sum().sum()
        location = address.get()
        xcol= xcolumn.get().strip()
        ycol = ycolumn.get().strip()
        filenaam = filename.get()
        col_val.set(columns)
        row_val.set(rows)
        null_val.set(nul)
        plt.plot(df[xcol],df[ycol])
        plt.xlabel(xcol)
        plt.ylabel(ycol)
        plt.savefig(filenaam)
        return 0
    else:
        print('Please fill the entry boxes \n')

    
    
def scatter():
    df = pd.read_csv(address.get())
    rows,columns = df.shape
    nul = df.isnull().sum().sum()
    location = address.get()
    xcol= xcolumn.get().strip()
    ycol = ycolumn.get().strip()
    filenaam = filename.get()
    col_val.set(columns)
    row_val.set(rows)
    null_val.set(nul)
    plt.scatter(df[xcol],df[ycol])
    plt.xlabel(xcol)
    plt.ylabel(ycol)
    plt.savefig(filenaam)
    return 0
    
    
def bar():
    df = pd.read_csv(address.get())
    rows,columns = df.shape
    nul = df.isnull().sum().sum()
    location = address.get()
    xcol= xcolumn.get().strip()
    ycol = ycolumn.get().strip()
    filenaam = filename.get()
    col_val.set(columns)
    row_val.set(rows)
    null_val.set(nul)
    plt.bar(df[xcol],df[ycol])
    plt.xlabel(xcol)
    plt.ylabel(ycol)
    plt.savefig(filenaam)
    return 0
    
    
def pie():
    df = pd.read_csv(address.get())
    rows,columns = df.shape
    nul = df.isnull().sum().sum()
    location = address.get()
    xcol= xcolumn.get().strip()
    ycol = ycolumn.get().strip()
    filenaam = filename.get()
    col_val.set(columns)
    row_val.set(rows)
    null_val.set(nul)
    plt.pie(df[xcol],df[ycol])
    plt.xlabel(xcol)
    plt.ylabel(ycol)
    plt.savefig(filenaam)
    return 0

def reset():
    return 0
    

    


headlabel_1= Label(win, text = 'Welcome to the graphical representation of your data', 
              fg = 'black' , font = ('arial',15,'bold')).pack()

headlabel_2= Label(win, text = 'This GUI used only for numerical computation("not for string values")', 
              fg = 'black' , font = ('arial',10,'bold'))
headlabel_2.place(x=10, y = 40)

headlabel_3= Label(win, text = 'Your chart file will save at that location ,where from u operate your GUI file', 
              fg = 'black' , font = ('arial',10,'bold'))
headlabel_3.place(x=700, y = 40)



label = Label(win, text = 'file Address/location', font = (30))
label.place(x=250,y=100)
addEntry = Entry(win, textvar= address, width = 60, border = 5, bg = 'cyan')
addEntry.place(x= 400, y= 100)

submit_button = Button(win, text='submit', width = 7, command = submit, bg='green', fg='yellow')
submit_button.place(x= 770, y= 100)

col_label = Label(win, text = 'columns names :',font = (10))
col_label.place(x=30,y=200)


label = Label(win, text = 'column("X axis")', font = (30))
label.place(x=30,y=300)
addEntry = Entry(win, textvar= xcolumn, width =30, border = 5, bg = 'cyan')
addEntry.place(x= 200, y= 300)

label = Label(win, text = 'column("Y axis")', font = (30))
label.place(x=30,y=400)
addEntry = Entry(win, textvar= ycolumn, width = 30, border = 5, bg = 'cyan')
addEntry.place(x= 200, y= 400)

label = Label(win, text = 'Save file', font = (30))
label.place(x=30,y=500)
addEntry = Entry(win, textvar= filename, width = 30, border = 5, bg = 'cyan')
addEntry.place(x= 200, y= 500)


addEntry = Entry(win, textvar= col_val, width = 30, border = 5)
addEntry.place(x= 800, y= 300)
label = Label(win, text = 'No. of columns', font = (30))
label.place(x=1020,y=300)

addEntry = Entry(win, textvar= row_val, width = 30, border = 5)
addEntry.place(x= 800, y= 400)
label = Label(win, text = 'No. of rows', font = (30))
label.place(x=1020,y=400)

addEntry = Entry(win, textvar= null_val, width = 30, border = 5)
addEntry.place(x= 800, y= 500)
label = Label(win, text = 'No. of null values', font = (30))
label.place(x=1020,y=500)





button = Button(win, text='line plot', width = 20, command = line, bg='red', fg='black', font = 15)
button.place(x= 200, y= 600)

button = Button(win, text='scatter', width = 20, command = scatter, bg='red', fg='black', font = 15)
button.place(x= 400, y= 600)

button = Button(win, text='bar', width = 20, command = bar, bg='red', fg='black', font = 15)
button.place(x= 600, y= 600)

button = Button(win, text='pie', width = 20, command = pie, bg='red', fg='black', font = 15)
button.place(x= 800, y= 600)

button = Button(win, text='reset', width = 20, command = reset, bg='red', fg='black', font = 15)
button.place(x= 1000, y= 600)






win.mainloop()
```
